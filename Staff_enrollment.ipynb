```python
import string
import random


def user_info(): 
    first_name = input('Enter Your FirstName: \n')
    last_name = input('Enter Your LastName: \n')
    email = input('Enter Email Address: \n')
    
    info = [first_name, last_name, email]
    return(info)

def passphrase(info):
    combo = str(info[0][:2] + info[1][-2:])
    random_length = 5
    auto_gen = ''.join(random.choice(string.ascii_lowercase) for r in range(random_length)) + combo
    return (auto_gen)

def random_id():
    mix = str(info[0][0] + info[1][0])
    ran = random.randint(1,99)
    staff_id = mix + str(ran)
    return(staff_id)

employee_stat = True
employee_db = []

while employee_stat:
    
    info = user_info()
    auto_gen = passphrase(info)
    staff_id = random_id()
    print(f'Your Autogenerated Passphrase \n {auto_gen} \n Staff Id \n {staff_id}')
    
    staff_choice = input('Change Default Passphrase? Y or N: \n')
    
    passphrase_loop = True
    
    while passphrase_loop:
        if staff_choice.upper() == 'N':
            info.append(auto_gen)
            employee_db.append(info)
            break
        
        else:
            new_passphrase = input('Enter New Passphrase, Longer than or Equal to 7 Characters: \n')
            
            if len(new_passphrase) < 7:
                print ('âš  Must be ATLEAST 7 Characters!')
            else:
                print ('Passphrase Looks Good ðŸ˜€')
                
                info.append(auto_gen)
                employee_db.append(info)
                break
                
    new_enrollment = input('Enroll New Staff? Y/N \n')
    
    if new_enrollment.upper() == 'N':
        
        employee_stat = False
        for data in employee_db:
            print(data)
            
    else:
        employee_stat = True
            
    
```
